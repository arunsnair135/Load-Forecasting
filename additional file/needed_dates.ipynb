```python
from datetime import date, timedelta

start_date = date(2021, 9, 1)
end_date = date(2022, 12, 31)
delta = timedelta(days=1)
needed_dates=[]
while start_date <= end_date:
    needed_dates.append(start_date.strftime("%Y-%m-%d"))
    start_date += delta
```

```python
dataset=pd.DataFrame(needed_dates,columns=['Date'])
```
